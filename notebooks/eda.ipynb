```python
top_city = df["City"].value_counts().head(10).index

city_df = df[df["City"].isin(top_city)]

plt.figure(figsize=(12,5))

sns.countplot(
    data=city_df,
    x="City",
    hue="Churn Label"
)

plt.xticks(rotation=45)

plt.show()
```
